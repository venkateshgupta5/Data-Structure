```python
class Node:
    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key
        
def PreorderTraversal(root):
        if root:
            print " "
            print root.val
            if root.left is not None:
                print "Left Node is",root.left.val
            else:
                print "Left is Null Found"
            print "---------->>> Root Left is called"
            PreorderTraversal(root.left)
            if root.right is not None:
                print "Right Node is",root.right.val
            else:
                print "Right is Null Found"
                
            
            print "---------->>> Root Right is called"
            PreorderTraversal(root.right)
                
root = Node('A')
root.left = Node('B')
root.right = Node('C')
root.right.left = Node('X')
root.right.right = Node('Y')
root.left.left = Node('D')
root.left.right = Node('E')
root.left.right.left = Node('F')

PreorderTraversal(root)
```
